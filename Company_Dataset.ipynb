```python
import matplotlib.pyplot as plt
```

```python
sorted_df = df.sort_values("Revenue_Lakhs")

plt.figure()
plt.plot(sorted_df["Revenue_Lakhs"].values)
plt.xlabel("Company Index")
plt.ylabel("Revenue (Lakhs)")
plt.title("Revenue Trend Across Companies")
plt.show()
```

```python
city_counts = df["City"].value_counts()

plt.figure()
plt.bar(city_counts.index, city_counts.values)
plt.xlabel("City")
plt.ylabel("Number of Companies")
plt.title("Companies per City (Bar Chart)")
plt.show()
```

```python
city_employees = df.groupby("City")["Employees"].mean()

plt.figure()
plt.bar(city_employees.index, city_employees.values)
plt.xlabel("City")
plt.ylabel("Average Employees")
plt.title("Average Employees per City")
plt.show()
```

```python
city_avg_revenue = df.groupby("City")["Revenue_Lakhs"].mean().sort_values()

plt.figure()
plt.plot(city_avg_revenue.index, city_avg_revenue.values)
plt.xlabel("City")
plt.ylabel("Average Revenue (Lakhs)")
plt.title("Average Revenue by City (Line Chart)")
plt.show()
```

```python
industry_avg_emp = df.groupby("Industry")["Employees"].mean()

plt.figure()
plt.plot(industry_avg_emp.index, industry_avg_emp.values)
plt.xlabel("Industry")
plt.ylabel("Average Employees")
plt.title("Average Employees by Industry")
plt.xticks(rotation=45)
plt.show()
```
